```python
from datetime import datetime

def calculate_import_tax(df: pd.DataFrame, make: str, model: str, year_manufacture: int):
    """
    Calculates the estimated total import tax payable for a vehicle in Kenya
    based on the cleaned CRSP DataFrame and user-provided details.
    """
    current_year = datetime.now().year 

    # Filter for exact make and model (case-insensitive)
    vehicle = df[
        (df['Make'].str.lower() == make.lower()) &
        (df['Model'].str.lower() == model.lower())
    ]

    if vehicle.empty:
        return "Vehicle not found in the CRSP list. Please check the make and model spelling."

    if vehicle.shape[0] > 1:
        return (f"Multiple entries found for {make} {model}. "
                "Consider refining the search or checking the dataset for variants.")

    row = vehicle.iloc[0]
    crsp = row['crsp_ksh']
    engine_cc = row['Engine_capacity']  # float or NaN for EVs/trucks
    fuel = row.get('Fuel', None)  # e.g., 'ELECTRIC', 'PETROL', 'DIESEL', 'HYBRID'

    # Calculate vehicle age
    age = current_year - year_manufacture
    if age < 0:
        return "Year of manufacture cannot be in the future."
    if age > 8:
        return "Vehicles older than 8 years cannot be imported into Kenya."

    # Depreciation percentage based on age (as per your provided rules)
    if age < 1:
        depreciation_pct = 0
    elif age <= 2:
        depreciation_pct = 20
    elif age <= 3:
        depreciation_pct = 30
    elif age <= 4:
        depreciation_pct = 40
    elif age <= 5:
        depreciation_pct = 50
    elif age <= 6:
        depreciation_pct = 55
    elif age <= 7:
        depreciation_pct = 60
    elif age <= 8:
        depreciation_pct = 65
    else:
        depreciation_pct = 65  # capped

    # True value = CRSP after depreciation
    true_value = crsp * (1 - depreciation_pct / 100)

    # Import Duty: 35% of true value
    import_duty = 0.35 * true_value

    # Excise Duty base = true_value + import_duty
    excise_base = true_value + import_duty

    # Determine excise duty rate
    if fuel == 'ELECTRIC':
        excise_rate = 0.10  # 10% for 100% electric vehicles
    elif pd.isna(engine_cc):
        # For heavy trucks or cases with no engine_cc, assume higher rate or default
        excise_rate = 0.35  # conservative; adjust if known
    elif fuel == 'DIESEL':
        if engine_cc > 2500:
            excise_rate = 0.35
        elif engine_cc > 1500:
            excise_rate = 0.25
        else:
            excise_rate = 0.20
    else:  # Petrol or Hybrid
        if engine_cc > 3000:
            excise_rate = 0.35
        elif engine_cc > 1500:
            excise_rate = 0.25
        else:
            excise_rate = 0.20

    excise_duty = excise_rate * excise_base

    # VAT base = true_value + import_duty + excise_duty
    vat_base = true_value + import_duty + excise_duty
    vat = 0.16 * vat_base

    # Railway Development Levy (RDL): 2% of true value (no compounding)
    rdl = 0.02 * true_value

    # Import Declaration Fee (IDF): 3.5% of true value (no compounding)
    idf = 0.035 * true_value

    # Total tax
    total_tax = import_duty + excise_duty + vat + rdl + idf

    # Breakdown for transparency
    breakdown = {
        "CRSP (KES)": round(crsp, 2),
        "Age (years)": age,
        "Depreciation (%)": depreciation_pct,
        "True Value (KES)": round(true_value, 2),
        "Import Duty (35%)": round(import_duty, 2),
        "Excise Duty": round(excise_duty, 2),
        "VAT (16%)": round(vat, 2),
        "Railway Development Levy (2%)": round(rdl, 2),
        "Import Declaration Fee (3.5%)": round(idf, 2),
        "Total Estimated Tax Payable (KES)": round(total_tax, 2)
    }

    return breakdown
```

```python
# Lets see if our function works 
result = calculate_import_tax(df, 'Peugeot', '3008 GT LINE', 2020)

if isinstance(result, str):
    print(result)
else:
    print("Import Tax Breakdown\n" + "="*40)
    for key, value in result.items():
        print(f"{key}: {value}")
```
